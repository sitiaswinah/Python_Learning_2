## Day 4

```python
# misalnya kita akan memanggil column Age yang value nya kurang dari sama dengan 50

df[~(df['Age']>50)]
```

```python
# misalnya kita akan memanggil column Age yang value nya lebihlebih dari 50 dan colmn Sex yang male

df[(df['Age']>50) & (df['Sex']=='male')]
```

```python
# misalnya kita akan memanggil column Age yang value nya lebih dari 50 atau column Sex yang male

df[(df['Age']>50) | (df['Sex']=='male')]
```

```python
# misalnya kita akan memanggil column Age yang value nya antara 20 dan 40

df[((df['Age']>20) & (df['Age']<40))]
```
